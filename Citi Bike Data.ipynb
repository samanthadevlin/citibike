```python
import pandas as pd
import numpy as np
```

```python
# File to Load
file_to_load = "Resources/citibike.csv"

# Read File and store into Pandas data frame
bike_data = pd.read_csv(file_to_load)
```

```python
bike_data.head()
```

```python
# show trip duration on bikeid 
bike_id = bike_data[["tripduration", "bikeid"]]
bike_id.head()
```

```python
bike_id.to_csv("Resources/bikeid.csv",index=False)
```

```python
#birth year start and stop time, intersting to see age and time
birth_time = bike_data[["birth year","starttime","stoptime"]]
birth_time
```

```python
birth_time.to_csv("Resources/birth_time.csv",index=False)
```

```python
#Gender (Zero=unknown; 1=male; 2=female) 
# I want to see how many male and female riders there are
gender = bike_data[["gender"]]
gender
```

```python
count=gender["gender"].value_counts()
count
```

```python
total_count=[count[0],count[1],count[2]]
total_count
```

```python
percent = np.round((count[0]/gender.shape[0]*100),2), np.round((count[1]/gender.shape[0]*100),2),np.round((count[2]/gender.shape[0]*100),2)
percent
```

```python
#Create Dataframe will total counts and percentages
#total count of riders and their percentages 
gender_bike=pd.DataFrame({"Percentage of Players": percent ,"Total Count": total_count })
gender_bike.index=(["Male 1", "Female 2", "Other/Non-Disclosed 0"])
gender_bike
```

```python
birth_time.to_csv("Resources/gender_bike.csv",index=False)
```

```python
top_start_stations = bike_data[["start station name"]]
top_start_stations
```

```python
top_start_stations = bike_data.groupby(["start station name"]).size().reset_index(name="counts")
top_start_stations.sort_values("counts", ascending=False).head(10)
```

```python
top_end_stations = bike_data[["end station name"]]
top_end_stations
```

```python
top_end_stations = bike_data.groupby(["end station name"]).size().reset_index(name="counts")
top_end_stations.sort_values("counts", ascending=False).head(10)
```

```python
top_start_stations.to_csv("Resources/top_start_stations.csv",index=False)
```

```python
top_end_stations.to_csv("Resources/top_end_stations.csv",index=False)
```
